```python
from pathlib import Path

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

resource_dir = Path("./")

# Read in the full CSV file from ZIP file in 'resources' directory
data = pd.read_csv(
    Path(resource_dir, 'dataset_preprocess1.CSV'),
    sep=',',
    header=None)

col_names = []

with open(Path(resource_dir, 'col_names_preprocess.txt')) as cols_fp:
    for line_num, name in enumerate(cols_fp):
        col_names.append(name.rstrip())

# Set the column headers to the names from the Wireshark frame
data.columns = col_names
```

```python
data.head()
```

```python
###MEMBUKA FILE VarHEx####
import re
f = open("./dataset/VarBukanINT.txt", "r")
VarHex = []
for i in f:
    VarHex.append(re.sub("\n",'',i))
f.close()
print(VarHex)
```

```python
###MENGHILANGKAN KOLOM YANG DULUNYA BUKAN INT####
for i in VarHex:
    data.drop(columns=[i], inplace=True)
```

```python
###DROP KOLOM OBJECT###
data.drop(columns=["wlan.ba.bm", "wlan_mgt.fixed.current_ap"], inplace=True)
```

```python
###DROP KOLOM OBJECT###
data.drop(columns=["wlan.qos.buf_state_indicated"], inplace=True)
```

```python
###DROP KOLOM YANG NILAI MAX = MIN #####
for i in data:
    if i == "class":
        continue
    if max(data[i]) == min(data[i]):
        print(i)
        data.drop(columns=[i], inplace=True)
```

```python
###DROP KOLOM OBJECT###
for x in data:
    if data[x].dtypes == "object":
        data.drop(columns=x, inplace=True)
```
