```python
from pathlib import Path

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

resource_dir = Path("./")

# Read in the full CSV file from ZIP file in 'resources' directory
x = pd.read_csv(
    Path(resource_dir, 'x_normalized1.CSV'),
    sep=',',
    header=None)

col_names = []

with open(Path(resource_dir, 'x_col_names_prenormalized.txt')) as cols_fp:
    for line_num, name in enumerate(cols_fp):
        col_names.append(name.rstrip())

# Set the column headers to the names from the Wireshark frame
x.columns = col_names
```

```python
y = pd.read_csv(
    Path(resource_dir, 'y_normalized.CSV'),
    sep=',',
    header=None)

ycol_names = ["class"]

# Set the column headers to the names from the Wireshark frame
y.columns = ycol_names
```

```python
temp = []
for i in y["class"]:
    if i != 0:
        i = 1
    temp.append(i)
```

```python
Y = pd.DataFrame(temp)
Y.columns = ["class"]
```

```python
a0 = 0
a1 = 0
a2 = 0
a3 = 0
a4 = 0
a5 = 0
a6 = 0
a7 = 0
a8 = 0
a9 = 0
for i in Y["class"]:
    if i == 0:
        a0 += 1
    elif i == 1:
        a1 += 1
    elif i == 2:
        a2 += 1
    elif i == 3:
        a3 += 1
    elif i == 4:
        a4 += 1
    elif i == 5:
        a5 += 1
    elif i == 6:
        a6 += 1
    elif i == 7:
        a7 += 1
    elif i == 8:
        a8 += 1
    elif i == 9:
        a9 += 1
print("0 =",a0)
print("1 =",a1)
print("2 =",a2)
print("3 =",a3)
print("4 =",a4)
print("5 =",a5)
print("6 =",a6)
print("7 =",a7)
print("8 =",a8)
print("9 =",a9)
```

```python
####SAMPLING MAJORITY####
# Undersample and plot imbalanced dataset with One-Sided Selection
from imblearn.under_sampling import OneSidedSelection

# define the undersampling method
undersample = OneSidedSelection(sampling_strategy='majority', n_neighbors=1, n_seeds_S=200, random_state=42)
# transform the dataset
x_res_maj, y_res_maj = undersample.fit_resample(x, Y)

# scatter plot of examples by class label
#for label, _ in counter.items():
#	row_ix = where(y == label)[0]
#	pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
#pyplot.legend()
#pyplot.show()
```
